```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 20, 1001)
```

```python
y = np.sin(x)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, "-")

ax.set(xlabel="x", ylabel="y")
ax.grid(True)
plt.show()
```
